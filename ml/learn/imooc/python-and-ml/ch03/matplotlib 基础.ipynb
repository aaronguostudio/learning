```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
print("siny is > ", siny)
cosy = np.cos(x)
print("cosy is > ", cosy)
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="sin(x)")
# plt.xlim(-5, 15)
# plt.ylim(0.02, 1)
plt.axis([-1, 11, -2, 2])
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Hello world from ML")
plt.show()
```

## Scatter plot

```python
plt.scatter(x, siny)
plt.scatter(x, cosy)
```
